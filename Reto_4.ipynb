# Conclusiones

- A partir de los resultados obtenidos, se observa que el modelo SwinIR mantiene una consistencia notable en la reconstrucción de las imágenes, logrando un PSNR promedio de 16.53 dB. Aunque numéricamente los valores de PSNR puedan parecer bajos en una escala absoluta, es importante considerar que el nivel de ruido aplicado (Noise 50) es bastante alto. Esto demuestra que la arquitectura basada en Transformers es capaz de rescatar información estructural incluso en condiciones de degradación severa.
- Un hallazgo interesante en las pruebas es la diferencia entre las métricas:

-     El mejor PSNR se obtuvo en la prueba 0_1536 (16.72 dB), lo que indica una mayor fidelidad en cuanto al error cuadrático medio de los píxeles.

-     Sin embargo, el mejor SSIM se presentó en la prueba 0_512 (0.5124). Esto sugiere que, en resoluciones más bajas, el modelo es más eficiente preservando la estructura y texturas visibles para el ojo humano, a pesar de que el valor de ruido sea ligeramente superior.
- El valor promedio de SSIM (0.4532) refleja la dificultad técnica de procesar imágenes con un nivel de ruido de 50. Se concluye que, si bien el modelo logra "limpiar" la imagen, la pérdida de detalles finos es inevitable bajo estas condiciones. El equilibrio entre el PSNR y el SSIM nos muestra que no siempre la imagen con "menos ruido matemático" es la que mejor se ve visualmente, resaltando la importancia de usar ambas métricas para evaluar modelos de Deep Learning.
